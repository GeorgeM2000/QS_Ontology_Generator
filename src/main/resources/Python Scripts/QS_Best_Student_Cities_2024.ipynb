```python
import pandas as pd

# Initialize lists to store data
ranks = []
cities = []
countries = []
scores = []

# Open the text file
with open('QS_Best_Student_Cities_2024.txt', 'r') as file:
    lines = file.readlines()

# Loop through the lines and extract information
for i in range(0, len(lines), 4):
    rank = lines[i].strip()
    city = lines[i+1].strip()
    country = lines[i+2].strip()
    score = float(lines[i+3].strip())

    # Append data to lists
    ranks.append(rank)
    cities.append(city)
    countries.append(country)
    scores.append(score)

# Create a DataFrame
df = pd.DataFrame({
    'Rank': ranks,
    'City': cities,
    'Country': countries,
    'Score': scores
})

for index, row in df.iterrows():
  print(f'{row["Rank"]} | {row["City"]} | {row["Country"]} | {row["Score"]}\n')
```

```python
df.to_excel('QS_Best_Student_Cities_2024.xlsx', index=False)
```
