```python
def extract_inner_tables_as_dataframes(file_path):
    # Load the Excel workbook
    wb = load_workbook(filename=file_path, read_only=True)

    # Initialize a list to store the DataFrames for each sheet
    dataframes = []

    # Iterate over each sheet in the workbook
    for sheet_name in wb.sheetnames:
        # Access the current sheet
        sheet = wb[sheet_name]

        # Extract data from the sheet and store it in a list of lists
        table_data = []
        for row in sheet.iter_rows(values_only=True):
            table_data.append(row)

        # Create a DataFrame from the table data
        df = pd.DataFrame(table_data)

        # Add the DataFrame to the list
        dataframes.append(df)

    # Close the workbook
    wb.close()

    return dataframes

dataframes = extract_inner_tables_as_dataframes("2024 QS WUR by Subject - Public Results (for qs.com)_56.xlsx")
dataframes = dataframes[1:]
```

```python
print(f'Number of Dataframes/Subjects: {len(dataframes)}')
```

```python
# Get the names of the subjects
subject_names = []
for df in dataframes:
  subject_names.append(df.iloc[1][0])
```

```python
print(f'Number of subjects: {len(subject_names)}')
```

```python
for subject in subject_names:
  print(subject)
```

```python
def process_dataframes(dataframes):
    for df in dataframes:
        # Initialize variables to store row and column indices
        row_index = None
        col_index = None

        # Search for the values "2024" and "2023" in the first and second columns
        for index, row in df.iterrows():
            if "2024" == row[0]:
                row_index = index

            # If both values are found, exit the loop
            if row_index is not None:
                break

        # Search for the value "None" in the same row
        if row_index is not None:
            for index, value in enumerate(df.iloc[row_index]):
                if value is None:
                    col_index = index - 1
                    break

        # Delete rows above the marked row
        if row_index is not None:
            df.drop(df.index[:row_index], inplace=True)

        # Delete columns to the right of the marked column
        if col_index is not None:
            df.drop(df.columns[col_index + 1:], axis=1, inplace=True)

# Example usage
process_dataframes(dataframes)
```

# ***Normalize the Universities, Countries, Cities and Subjects***

```python
wur_subject = pd.read_csv("QS 2024 WUR by Subject Filtered.csv")
wur = pd.read_csv("QS 2024 World University Rankings.csv")
student_cities = pd.read_csv("QS Best Student Cities 2024 Filtered.csv")
```

```python
# Define a function to apply transformations to a single value
def process_value(value):
    # Replace whitespaces with underscore
    modified_value = re.sub(r'\s', '_', value)
    # Remove symbols
    modified_value = re.sub(r'[^a-zA-Z0-9_]', '', modified_value)
    return modified_value
```

```python
wur_columns = ['institution', 'location']

for column in wur_columns:
    wur[column] = wur[column].apply(process_value)

wur = pd.DataFrame(wur)

wur.head()
```

```python
wur_subject_columns = ['Institution', 'Location', 'Subject']

for column in wur_subject_columns:
    wur_subject[column] = wur_subject[column].apply(process_value)

wur_subject = pd.DataFrame(wur_subject)

wur_subject.head()
```

```python
wur_columns = ['City', 'Country']

for column in wur_columns:
    student_cities[column] = student_cities[column].apply(process_value)

student_cities = pd.DataFrame(student_cities)

student_cities.head()
```

```python
student_cities.to_csv("QS Best Student Cities 2024 Filtered.csv", index=False)
wur_subject.to_csv("QS 2024 WUR by Subject Filtered.csv", index=False)
wur.to_csv("QS 2024 World University Rankings Filtered.csv", index=False)
```
